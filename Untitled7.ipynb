```python
import matplotlib.pyplot as plt

def quadratic_temperature_model(day, a, b, c):
    temperature = a * day**2 + b * day + c
    return temperature

def main():
    # Hard-coded coefficients for the quadratic equation: ax^2 + bx + c
    a = float(input("Enter the quadratic coefficient (a): "))
    b = float(input("Enter the linear coefficient (b): "))
    c = float(input("Enter the constant term (c): "))

    # Get user input for the number of days to model
    num_days = int(input("Enter the number of days to model: "))

    days = list(range(0, num_days + 1))
    temperatures = [quadratic_temperature_model(day, a, b, c) for day in days]

    # Plotting
    plt.plot(days, temperatures, marker='o', label=f'Temperature: {a}x^2 + {b}x + {c}')
    plt.title("Weather Modeling")
    plt.xlabel("Day")
    plt.ylabel("Temperature (°C)")
    plt.legend()
    plt.show()

main()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def quadratic_weather_model(time, a, b, c):
    return a * (time ** 2) + b * time + c

def main():
    print("Quadratic Weather Modeling")
    print("---------------------------")

    file_path = 'task3.csv'  # path to your CSV

    try:
        # Read CSV, skip blank or malformed rows
        df = pd.read_csv(file_path, header=None)
        df.columns = ['a', 'b', 'c']

        # Convert to numeric and drop NaN rows
        df = df.apply(pd.to_numeric, errors='coerce').dropna()

        time_values = list(range(0, 11))
        plt.figure(figsize=(8, 6))

        for _, row in df.iterrows():
            a, b, c = row['a'], row['b'], row['c']
            temperature_values = [quadratic_weather_model(t, a, b, c) for t in time_values]
            plt.plot(time_values, temperature_values, marker='o', linestyle='-', label=f'a={a}, b={b}, c={c}')

        plt.title("Temperature Variation Over Time")
        plt.xlabel("Time")
        plt.ylabel("Temperature")
        plt.grid(True)
        plt.xticks(range(0, 11, 1))
        plt.legend()
        plt.show()

    except FileNotFoundError:
        print(f"File not found. Please make sure '{file_path}' exists.")

main()
```
